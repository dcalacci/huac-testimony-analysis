```python
def get_key(mention, l):
    "second argument is the chunked list."
    mention = mention.lower()
    for chunk in l:
        if mention in chunk:
            return l.index(chunk)
```

```python
disambiguated_summary_graph = {}
for snitch, accused in named_graph.items():
    snitch_key = get_key(snitch, disambiguated_names)
    accused = [get_key(a, disambiguated_names) for a in accused]
    disambiguated_summary_graph[snitch_key] = accused
```

So actually, there are lots of rows that are just empty lists that aren't np.nan. Let's change them to np.nan if they are empty lists:

```python
def empty(l):
    if type(l) == list:
        return all(empty(x) for x in l)
    return False
    
def convert_empty_to_nan(l):
    if empty(l):
        return np.nan
    return l
```

```python
speechacts['mention_list_by_sentence_with_anaphora'] = speechacts['mention_list_by_sentence_with_anaphora'].apply(convert_empty_to_nan)
speechacts['mention_list_by_sentence_without_anaphora'] = speechacts['mention_list_by_sentence_without_anaphora'].apply(convert_empty_to_nan)
```

### Disambiguate speakers and mentions

When searching for a key for a name, need to apply normalize_name before checking if its in the list.

```python
def normalize_name(s):
    return ' '.join(s.lower().split())
```

get all the speaker and mention names

```python
speakers = speechacts_with_mentions['speaker']

def flatten_list(l):
    return [item for sublist in l for item in sublist]


mentions = list(speechacts_with_mentions['mention_list_for_speechact_without_anaphora'].dropna())
mentions = flatten_list(mentions)
mentions_only_anaphora = flatten_list(list(speechacts_with_mentions['mention_list_by_sentence_with_anaphora'].dropna().apply(flatten_list)))

all_mentions = mentions_only_anaphora + mentions

# all names in mentions and speakers
all_names = all_mentions + list(speakers)

all_names = [normalize_name(s) for s in all_names]
```

```python
disambiguated_transcript_names = text_utils.chunk_list(set(all_names))
```

```python
disambiguated_transcript_names
```

### Create a field for disambiguated mentions by sentence & by speechact.

```python
def convert_names_to_ids(l):
    "convert list of lists of names to lists of lists of keys"
    def get_hearing_key(s):
        return get_key(normalize_name(s), disambiguated_transcript_names)
    if type(l) == float:
        return l
    return map(lambda x: map(lambda y: get_hearing_key(y), x), l)

def convert_list_of_names_to_ids(l):
    return [get_key(normalize_name(s), disambiguated_transcript_names) for s in l]
```

```python
speechacts_with_mentions['disambig_mention_list_by_sentence_with_anaphora'] = speechacts_with_mentions['mention_list_by_sentence_with_anaphora'].apply(convert_names_to_ids)
```

```python
speechacts_with_mentions['disambig_mention_list_by_sentence_without_anaphora'] = speechacts_with_mentions['mention_list_by_sentence_without_anaphora'].apply(convert_names_to_ids)
```

```python
speechacts_with_mentions['disambig_mention_list_for_speechact_without_anaphora'] = speechacts_with_mentions['mention_list_for_speechact_without_anaphora'].apply(convert_list_of_names_to_ids)
```

```python
speechacts_with_mentions['disambig_speaker'] = speechacts_with_mentions['speaker'].apply(lambda s: get_key(normalize_name(s), disambiguated_transcript_names))
```
